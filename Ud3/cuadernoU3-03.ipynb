En este mapa de bits, los caracteres de espacio representan un espacio vacío, y todos los demás caracteres se reemplazan por caracteres en el mensaje del usuario. El mapa de bits proporcionado se asemeja a un mapa del mundo, pero puedes cambiar esto a cualquier imagen que desees. La simplicidad binaria del sistema de caracteres espacio-o-mensaje lo hace bueno para los principiantes. ¡Intenta experimentar con diferentes mensajes para ver cómo son los resultados!

## 3.2 Cómo funciona

En lugar de escribir individualmente cada carácter del patrón de mapa del mundo, puedes copiar y pegar todo desde https://inventwithpython.com/bitmapworld.txt. Una línea de 68 puntos en la parte superior e inferior del patrón actúa como una regla para ayudarte a alinearlo correctamente. Sin embargo, el programa seguirá funcionando si cometes errores con el patrón.

La llamada al método `bitmap.splitlines()` en la línea 43 devuelve una lista de cadenas, cada una de las cuales es una línea en la cadena multilínea `bitmap`. El uso de una cadena multilínea hace que el mapa de bits sea más fácil de editar con cualquier patrón que desees. El programa completa cualquier carácter que no sea espacio en el patrón, por lo que con los asteriscos, los puntos o cualquier otro caracter conseguimos lo mismo.

El código en la línea 51 `message[i % len(message)]` provoca la repetición del texto en `message`. Conforme `i` va incrementándose de 0 a un número mayor que `len(message)`, la expresión `i % len(message)` devuelve 0 otra vez. Esto hace que `message[i % len(message)]` repita los caracteres de `message` conforme `i` se va incrementando.


````python
"""Bitmap Message, by Al Sweigart al@inventwithpython.com
Displays a text message according to the provided bitmap image.
This code is available at https://nostarch.com/big-book-small-python-programming
Tags: tiny, beginner, artistic"""

import sys

# (!) Try changing this multiline string to any image you like:

# There are 68 periods along the top and bottom of this string:
# (You can also copy and paste this string from
# https://inventwithpython.com/bitmapworld.txt)
bitmap = """
....................................................................
   **************   *  *** **  *      ******************************
  ********************* ** ** *  * ****************************** *
 **      *****************       ******************************
          *************          **  * **** ** ************** *
           *********            *******   **************** * *
            ********           ***************************  *
   *        * **** ***         *************** ******  ** *
               ****  *         ***************   *** ***  *
                 ******         *************    **   **  *
                 ********        *************    *  ** ***
                   ********         ********          * *** ****
                   *********         ******  *        **** ** * **
                   *********         ****** * *           *** *   *
                     ******          ***** **             *****   *
                     *****            **** *            ********
                    *****             ****              *********
                    ****              **                 *******   *
                    ***                                       *    *
                    **     *                    *
...................................................................."""

print('Bitmap Message, by Al Sweigart al@inventwithpython.com')
print('Enter the message to display with the bitmap.')
message = input('> ')
if message == '':
    sys.exit()

# Loop over each line in the bitmap:
for line in bitmap.splitlines():
    # Loop over each character in the line:
    for i, bit in enumerate(line):
        if bit == ' ':
            # Print an empty space since there's a space in the bitmap:
            print(' ', end='')
        else:
            # Print a character from the message:
            print(message[i % len(message)], end='')
    print()  # Print a newline.
````
